```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
```

```python
df = pd.read_csv("geo_facts.csv")
df.info()
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
def get_relevance_score(location, sentence):
    embeddings = model.encode([location, sentence], convert_to_tensor=True)
    cosine_score = util.pytorch_cos_sim(embeddings[0], embeddings[1]).item()
    return cosine_score
```

```python
relevance_scores = []
for _, row in df.iterrows():
    location = row["location"]
    scores = [get_relevance_score(location, row[f"sentence {i}"]) for i in range(1, 6)]
    avg_score = sum(scores) / 5
    relevance_scores.append(avg_score)
```

```python
df["location_relevance_score"] = relevance_scores
df.to_csv("relevance_results.csv", index=False)
print(df[["location", "location_relevance_score"]].head())
```
